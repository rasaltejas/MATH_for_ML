```python
#Vector
import numpy as np
v = np.array([25, 30, 80])
print(v)
print(v.shape)
print(v.ndim)
```

```python
#Matrix
# Rows = students
# Columns = subjects
import numpy as np
students = np.array([
    [80, 50, 80],
    [80, 70, 90],
    [50, 60, 30]]
)
print(students)
print(students.shape)
print(students.ndim)  #used for check dimensions
```
